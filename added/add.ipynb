### การส้ราง class

```python
class Employee: #การสร้าง class
    pass

#ต้องการสร้าง object 
emp1 = Employee() #> เอาชื่อ class มาใช้สร้าง object >> การเรียกชใช้ class เเละใช้งานจริง
```

```python
class Employee:
    def details(self, name, salary, department) -> None: #การสร้าง method
        self.name = name
        self.salary = salary
        self.department = department

    def show_data(self):
        print(f"ชื่อพนักงาน : {self.name} \n เงินเดือน : {self.salary} \n เเผนก : {self.department}")

#ต้องการสร้าง object 
emp1 = Employee() 
emp1.details("kam" , 500000, "IT") # เรียกใช้งาน method details ผ่าน object emp1

obj1 = Employee()
obj1.details("wawa", 15000, "accounting")

cheetah = Employee()
cheetah.details("Cheetah", 20000, "CEO")

cheetah.show_data()
```

### การสร้าง constructure, destructure

```python
class Employee:

    def __init__(self, name, salary, department) -> None: #อันนี้เป็น private method
        # print("default constructor ==> เรียกใช้งานทันทีที่มีการสร้าง object")
        
    #self.attribute = เป็น public attribute สามารถเเก้ไขได้
        self.name = name
        self.salary = salary
        self.department = department
    
    #public method
    def show_data(self):
        print(f"ชื่อพนักงาน : {self.name} \n เงินเดือน : {self.salary} \n เเผนก : {self.department}")
    
    # def __del__(self):
    #     print("calling destructor ==> ทำงานตอนที่ object ถูกลบออกจากหน่วยความจำ")

kam = Employee("kam" , 500000, "IT") 

wawa = Employee("wawa", 15000, "accounting")

cheetah = Employee("Cheetah", 20000, "CEO")

kam.show_data()
wawa.show_data()
cheetah.show_data()
```

### Access modifiy

- Public
- Protecter
- Private

```python
class Employee:

    def __init__(self, name, salary, department) -> None: 
        
    #self.attribute เป็น protected attribute 
        self._name = name
        self._salary = salary
        self._department = department
    
    #Protected method
    def _show_data(self):
        print(f"ชื่อพนักงาน : {self._name} เงินเดือน : {self._salary} เเผนก : {self._department}")

kam = Employee("kam" , 500000, "IT") 
wawa = Employee("wawa", 15000, "accounting")
cheetah = Employee("Cheetah", 20000, "CEO")

kam._show_data()
wawa._show_data()
cheetah._show_data()
```
