```python
def reverse(data):
    # length = len(string)
    for index in range(len(data) - 1, -1, -1):
        yield data[index]
```

```python
x = iter(reverse([1,2,3,4,5]))
print(next(x))  # 5
print(next(x))  # 4
print(next(x))  # 3
print(next(x))  # 2
print(next(x))  # 1
# print(next(x))  # nodata so we get the error StopIteration
```

```python
def custom_generator(data):
    iterable = iter(data)
    for index in range(len(data)):
        yield index, iterable.__next__()
        index += 1

for item in custom_generator(['a','b','c']):
    print(item)
```

```python
class CustomIterator:
    def __init__(self, data):
        self.data = data
        self.index = -1

    def __iter__(self):
        return self # returns the iterator object itself

    def __next__(self):
        if self.index == len(self.data) - 1:
            raise StopIteration
        self.index += 1
        return self.index, self.data[self.index]
    
for item in CustomIterator(['a','b','c']):
    print(item)
```
