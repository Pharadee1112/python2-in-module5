7. try except else finally #ตัวควบคุม errorให้โปรแกรมไปต่อได้
ตัวสั่งให้เราจะทำไงให้เเก้ error ได้ 

```python
#ทำงานโดย จะทำ try ก่อนเสมอ ถ้าไม่เกิด error ขึ้น ถ้าเกิดจะเปลี่ยนไปเป็น except
#จะปริ้น except ทันทีเมื่อผิดก่อนหน้ามันผิด
try :
    print('hello')
except:
    print('world')
```

```python
try :
    print('hello')
    x = 2/0 
except:
    print('world')
#บรรทัดแรกปริ้น try ออกมา บรรทัดที่ 2 มัน zerodivison error ดังนั้นจึงปริ้น except
```

```python
try :
    print('hello')
    x = 2/0 
    print('my') #อันนี้ถูก เเต่บรรทัดบนมัน zerodivison error ดังนั้นมันจะข้ามไปปริ้น except
except:
    print('world')
```

ข้อระวังคือ except มันจะดักทุกตัวที่ error ดังนั้นอยากให้เป็นเฉพาะตัวที่ดัก เเนะนำให้ใส่ตัวที่ต้องการดัก

```python
try :
    print('hello')
    x = 2/0 
except ZeroDivisionError :
    print('world')
```

```python
try :
    print('hello')
    x = '2'+ 2
except ZeroDivisionError :
    print('world')
# miss type ตรง except -> typeerror  ถ้าเราไม่ดักอะไรเลย ก็จะรันต่อได้เหมือนเดิม
```

```python
i = int(input())
l = [1,2]
if i >= len(l):
    raise IndexError(f'รู้ไหมว่า index เริ่มที่ 0,มึงใส่ {i}เข้ามาอะ ไอ้ควย')
```

### assert = make sure that...

```python
import random

n = int(input())

assert 1 <= n < 10 #ถ้าใส่ตั้งเเต่ 10 นั้นคือจะให้เป็น assertionerror

if random.randint(1,10) == n:
    print('คุณถูกรางวัลนะเว้ยเพื่อลลล')
else :
    print('หวยแดกไหมละมั้ง สม')
```

```python
import random

n = int(input())

assert 1 <= n < 10  ,'ไม่มีหวยที่คูณเลือก เลือก 1-10 สิ อย่าโง่ไอ้บ้า'
#ถ้าใส่ตั้งเเต่ 10 นั้นคือจะให้เป็น assertionerror

if random.randint(1,11) == n:
    print('คุณถูกรางวัลนะเว้ยเพื่อลลล')
else :
    print('หวยแดกไหมละมั้ง สม')
```
