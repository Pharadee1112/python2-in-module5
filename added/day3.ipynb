```python
def gcd(x,y):#คือ หรม ที่ตัวมากที่สุดดังนั้นเราจะเอาตัวที่มากที่สุดเป็นตัวแรก
#ิอยากให้เร็วมากขึ้นก็เปลีี่ยน เป็น gcd_faster ได้ ถ้าเอา หรม ก็ใช้ min

#concept คือ การเอาผลลัพธ์ไปหาร x และ y แล้วหารเล้วได้ 0 ให้ย้อนกลับ s
    s = min(x,y)
    while True:
        if (x%s == 0) and (y%s == 0):
            return s
        else:
            s = s-1 #ถ้าไม่ก็คือส่งกลับมา เเล้วจะรันต่อไปเรื่อย ๆ gcd ที่น้อยที่สุดคือ 1 นั้นคือจุดที่มันหยุดอยู่เสมอ

import time

gcd(2,25) 
t1 = time.time()
print(gcd(221564,254))
t2 = time.time()
print(t2 - t1)
```

```python
def gcd_faster(x,y):
    s = min(x,y)
    while True:
        if (x%s == 0) and (y%s == 0):
            return s
        else:
            s = s-1

gcd(15,30)
gcd_faster(2,5468)
```
