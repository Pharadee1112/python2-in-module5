*args ดอกจันทร์อันเดียว

ปัญหาเกิดจาก ครน หรม หรือฟังก์ชัน เราสามารถรับกี่ตัวก็ได้ ไม่สำคัญว่าเป็นจำนวนเดียว

```python
def product(a,b,c): #อันนี้เราใส่ที่ตัวแปลให้กับจำนวนที่เราจะใส่เข้าไปเท่านั้น ดังนัน้เราต้องมี *args 
    result = a*b*c
    return result

product(6,8,1)
```

```python
def product(*args):
    print(args)

product(2)  #สังเกตว่าเมื่อเรา def args เเล้วมันออกมาเป็น , เสมอเลย 
```

ปัญหาเกิดจาก ครน หรม หรือฟังกชัน เราสามารถรับกี่ตัวก็ได้ ไม่ fix จำนวน input

def product คือการจำนวนทั้งหมดมันมาคูณกัน 

```python
def product (*args):
    result = 1 
    for number in args:
        result = result * number
    return result 
product (2,3,4) #เราจะเอา tuple เราจะเก็บไว้ในตัวเเปร number แล้วคูณด้วย 1 ,เเล้วก็ loop ไปเรื่อย ๆ ,มันก็จะคุณกันไปเรื่อย ๆ 
```

เราสามารถใช้ args คู่กับ parameter ได้ด้วย

```python
def product (a,*args): #เราเปลี่ยน 1 เป็นตัวแปล a = 2 เเล้วตัวต่อไปก็จะมาคูณเป็นปกติ
    result = a   
    for number in args:
        result = result * number
    return result 
product (2,3,4)
```

```python
def product (a, b,*args): 
    result = a
    for number in args:
        result = result * number
    result += b #เราเก็บตัวเเปร B ไว้บวกทีหลัง เพรามันไม่เกี่ยวข้องกับสมการเลย
    return result 

product (2,3,4,5,6) #เกิดจาก 2*4*5*6 เเล้วบอก 3(b)
```

```python
def calculate_something(a,b,c, *args , x,y):
    result = 1
    result += a+b+c 
    result *= args[1]
    result -= x*y 
    return result

calculate_something (1,2,3,5,6,8,10) 
```

```python
def calculate_something(a,b,c, *args ,x,y):
    result = 1
    result += a+b+c 
    result *= args[1] #อันนี้คือ args ของตำเเหน่งที่ 1
    result -= x*y  # keyword only argument ดังนั้นเราต้องมีตัวแปลให้ตัวเเต่ว่ามันก็เข้า args อยู่่ดี ดังนั้นเราต้องใส่ที่เป็นค่าของมันเลย คือ x = 11 , y = 12
    return result

calculate_something (1,2,3,5,6,8,10,x = 11 , y = 12) #args = คือจำนวนอะไรก็ได้ คือ 5 6 8 10
```

**kwargs มีดอกจันทร์ 2 อัน เเก้ปัญหากับ args 
ใช้เมื่อมี default parameter ไม่จำเป็นต้องกำจัดตัว (ตั้งเเต่ 0 ยัน infty)

```python
def greeting_student(**kwargs):
    # print(kwargs) #เราก็เอาไปใช้ได้เลย เช่น
    # print(kwargs['name1']) 

    msg = "สวัสดีตอนเช้า "
    for key in kwargs :
        msg += kwargs[key] + " " #ในเว้นวรรค

    return msg.strip() #มี .strip ก็ได้ เพราะว่ามันคือการตัดท้ายออกไป สังเกตว่ามันมีช่องว่างท้ายคำว่า Kaam , strip เพื่อตัดช่องว่างที่ท้ายประโยค
greeting_student(name1 = "win" ,  name2 = "kaam") #keyword argument ได้แบบไม่กำจัดเลย
print(type(greeting_student))
```

เอามารวมกัน 

```python
def calculate_sum(a,b,c,h=1 ,*args,d,e ,f=5, **kwargs): #ไม่ต้อง default parameter หลัง kwargs เพราะว่าเราสารมารถใส่ parameter ลงไปแบบไม่จำกัด กับ kwargs อยุ่เเล้ว 
# มี keyword only argument ที่ต้องใส่เข้าไปกับ d,e
# หน้า args คือตัวเลยที่เรากำหนดให้มันเป็น โดย h=6 มันจะเปลี่ยนค่าให้เอง
# เเละหน้า kwargs คือต้องเป็น default parameter 
    result = 0
    result += a+b+c
    result = h

    for number in args :
        result += number
    result  += d
    result += e
    result += f
    return result

calculate_sum(2,3,4,6,8,5,4,7,5,10,12,11,13,14,d=4,e=6) #ดังนั้นหลัง args เราต้องใส่ default parameter
#เเล้วเราก็จะนิยามมันใหม่ได้ ด้วยการใส่ค่าให้มันใหม่ เช่น f=7 มันก็จะเปลี่นค่าใหม่เลย
```

```python
def count_letter_a(s):
    kayyy= 0
    for character in s:
        if character == "a":
                kayyy += 1
    return kayyy
        
count_letter_a("kaaaaam")
```

```python
#หรืออออออออ
def count_letter_a(s : str): #อันนี้คือตนอที่เราไม่ได้เขียน : กำกับว่ามันคืออะไร
    #เเต่เราเขียนเเล้วว่า s:str มันจะรู้ว่าเป็นข้อมูลประเภทอะไร เเล้วเราสามารถเลือกหาฟังก์ชันออกมาใช้ได้เลย
    counter = s.count("a")
    return counter
count_letter_a("kammamamama")
```
