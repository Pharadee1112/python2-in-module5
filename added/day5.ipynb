function in function (nested function)

```python
x = 5

def f(): #เเละฟังก์ชัน g เป็นของ f
    global x
    x = 6 # local of f()
    
    def g():#local of g()
        x = 7
        print("Running function G")

    print(x) 

    g() #ใช้ฟังชัน g() ใน f()

# g() #ใช้ด้านนอก f ไม่ได้
f() #เรียกใช้
# print(x)
```

```python
x = 5

def f(): 
    # global x
    x = 6
    
    def g():
        global x 
        print(x) #ปริ้น x มันชี้ x = 5 (บรรทัดเเรก)
        x = 7

        # print(x) 
        print("Running function G") #(ปริ้นบรรทัดที่ 2)

    # print(x) 

    g() 

f()
# g() ใช้ไม่ได้เพราะ มันต้องอยู่ใน f
print(x) #(ปริ้นบรรทัดที่ 3)
```

###factorial

n! = n * (n-1)!

```python
def factorial(n : int):
    return n * factorial(n-1)

factorial(4)
```

```python
def factorial(n : int):
    if n == 1:
        return 1 #คือคืนค่าตัวสุดท้าย มันจะไม่คืนค่าตัวรีเทิร์นที่ 2 อะ 
    return n * factorial(n-1)

factorial(4) #อธิบายนะ >>>  n = 4 ไม่เข้าเงื่อนไขที่ 1 ก็จะไป return ที่ 2 ตอบ 24
             #ถ้า n = 1 เข้าเงื่อนไขอันเเรกก็จะ return กลับมา คือ 1 เพราะว่ามันคือ 1* factorial(1-1) = 0 ผิด ดังนั้น ต้อง return 1
factorial(1) 
```

$fibonacci$

$1,1,2,3,5,8,13,21,34,...$

$fibo(n) = fibo(n-1) + fibo(n-2)$

$fibo(1) = 1 \\
fibo(2) = 1\\
fibo(3) = 2 \\
fibo(4) = 3\\
fibo(5) = 5\\
...$

```python
def fibonacci(n : int):
    if n == 1:
        return 1
    if n == 2:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)

fibonacci(9)
```

```python
fibonacci(39) #ช้าเพราะว่า มันคำนวนทั้งหมด ยกกำลัง 39 รอบ
```

### dynamic programming

```python
def fibo(n: int):
    dp = [0 for i in range (n+1)]
    dp[0] = 1
    dp[1] = 1
    for i in range(2,n+1):
        dp[i] = dp[i-1] + dp[i-2]
    
    return dp[n-1]

fibo(1000)
```
