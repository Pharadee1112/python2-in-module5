```python
class Student:
    def __init__(self,age ,weight,height,bmi = None):
        self.age = age
        self.weight = weight
        self.height = height
        self.bmi = bmi
       

    def print_age(self):
        print(f'i am {self.age} years old.')

    def calculate_bmi(self):
        self.bmi = self.weight / (self.height/100) **2
        return self.bmi
    
    def evalute_bmi(self):
        assert self.bmi is not None , "BMI value is missing"
        
        if self.bmi < 18.5:
            return "underweight"
        
        if self.bmi < 22.9:
            return "okay"
        
        if self.bmi < 24.9:
            return "overweight"
        
        if self.bmi < 29.9:
            return "diabetes level 1"
        
        return "diabetes level 2"
```

```python
due = Student(age = 19, weight= 54, height= 155,bmi = None)

due.print_age()
due.calculate_bmi()
due.evalute_bmi()
```

```python
import random

class NumberGuessingGame:
    def __init__(self):
        self.number_to_guess = random.randint(1, 100)  # ‡∏™‡∏∏‡πà‡∏°‡∏´‡∏°‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 1 ‡∏ñ‡∏∂‡∏á 100
        self.attempts = 0  # ‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡πÄ‡∏Å‡πá‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏µ‡πà‡∏ó‡∏≤‡∏¢
        self.is_game_over = False  # ‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡πÄ‡∏ä‡πá‡∏Ñ‡∏™‡∏ñ‡∏≤‡∏ô‡∏∞‡πÄ‡∏Å‡∏°

    # ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏£‡∏¥‡πà‡∏°‡πÄ‡∏Å‡∏°
    def start_game(self):
        print("Welcome to the Number Guessing Game!")
        print("I have selected a number between 1 and 100. Try to guess it!")

        while not self.is_game_over:
            self.play_turn()

    # ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏•‡πà‡∏ô‡∏ó‡∏µ‡∏•‡∏∞‡∏£‡∏≠‡∏ö
    def play_turn(self):
        guess = input("Enter your guess: ")

        if not guess.isdigit():
            print("Please enter a valid number!")
            return  # ‡∏Å‡∏•‡∏±‡∏ö‡πÑ‡∏õ‡πÄ‡∏£‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏ú‡∏π‡πâ‡πÄ‡∏•‡πà‡∏ô‡∏Å‡∏£‡∏≠‡∏Å‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
        
        guess = int(guess)
        self.attempts += 1

        if guess < self.number_to_guess:
            print("Too low! Try again.")
        elif guess > self.number_to_guess:
            print("Too high! Try again.")
        else:
            self.is_game_over = True
            print(f"Congratulations! You guessed the number in {self.attempts} attempts.")

# ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡πÄ‡∏Å‡∏°
game = NumberGuessingGame()
game.start_game()
8
```
