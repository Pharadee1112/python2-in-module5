Create a Cricle class and intialize it with `radius`. Make two methods `get_area` and `get_circumference` inside this class. Try thinking about what properties should be of class and which should be of instances, which should be public and which should be private.

````python
import math

class Circle:
    
    def __init__(self, radius : int) -> None :
        self.radius = radius

    def get_area(self)-> float :
        return round(math.pi * (self.radius ** 2), 4)
    
    def get_circumference(self) -> float :
        return round(2 * math.pi * self.radius , 4)
    

wawa = Circle(radius = 12)

print(wawa.get_area())
print(wawa.get_circumference())
print(wawa.__init__(12))
````

Write a Python class named `Rectangle` constructed by a `length` and `width` and methods which will compute the area of a rectangle, and another method that calculate the perimeter.

````python
class Rectangle:
    
    def __init__(self, lenght : int, width : int) -> None :
        self.lenght = lenght
        self.width = width

    def calulate_ares(self) -> int:
        return self.lenght * self.width
    
    def calculate_perimeter(self) -> int:
        return (self.lenght * 2) + (self.width * 2)
    
cheetah = Rectangle(lenght=2, width=4)
print(cheetah.calculate_perimeter())
print(cheetah.calulate_ares())
````

Create a Clock class and initialize it with hours and minutes.
- Create a method `add_time()`, which should accept an argument – another `Clock` instance object – and adds it:
```python
clock1 = Clock(23, 30)
clock2 = Clock(14, 20)
clock1.add(clock2)
print(clock1.hours, clock1.minutes)
>>> 13, 50
```
- Create a method `display_time()` which should print the time.
- Create a method `display_total_minutes()`. E.g.- (1 hr 2 min) should display 62 minutes.

````python
class Clock:
    def __init__(self, hours : int, minutes : int):
        self.hours = hours
        self.minutes = minutes

    def add_time(self, another_clock):
        total_minutes = self.minutes + another_clock.minutes #50
        remain_hours = total_minutes // 60 #0
        self.minutes = total_minutes % 60 #50

        total_hours = self.hours + another_clock.hours + remain_hours #37
        self.hours = total_hours % 24 #13

    def display_time(self):
        return f"{self.hours}.{self.minutes}"
    
    def display_total_minutes(self):
        return (self.hours * 60) + self.minutes
    

clock1 = Clock(23, 30)
clock2 = Clock(14, 20)

clock1.add_time(clock2)

print(clock1.hours, clock1.minutes)

print(clock1.display_time())
print(clock1.display_total_minutes())
````

Create a Python class called `BankAccount` which represents a bank account, having as attributes: `accountNumber` (numeric type), `name` (name of the account owner as string type), `balance`.
- Create an `__init__` method with parameters: `account_number`, `name`, `balance`.
- Create a `put_money()` method which deposit money in and would raise an exception for a negative argument.
- Create a `withdraw()` method which withdraw money out and would raise an exception for a negative argument.
- Create an `apply_bank_fees()` method to apply the bank fees with a percentage of 5% of the `balance` amount, deduct the balance with the calculated fee.
- Create a `display()` method to display account details.

Try thinking about what properties should be of class and which should be of instances, which should be public and which should be private.

````python
class BankAccount:
    def __init__(self, account_number : int, name : str, balance : int):
        self.account_number = account_number
        self.name = name
        self._balance = balance

        if type(account_number) != int:
            raise TypeError("be int")

    def put_money(self, moneyin):
        if moneyin < 0 :
            raise ValueError("money can not less then 0")

        self._balance += moneyin

    def withdraw(self, moneyout):
        if moneyout > self._balance :
            raise ValueError("can not withdraw more than money account")

        self._balance = self._balance - moneyout

    def apply_bank_fees(self) :
        fees = (self._balance * 5) / 100
        self._balance -= fees

    def display(self):
        print(f"name acount : {self.name} \n number account : {self.account_number} \
              \n fee : 5% \n balance : {self._balance}")    
        
sun = BankAccount(112233, "Kam", 1000)

sun.put_money(200)
sun.withdraw(200)
sun.apply_bank_fees()

sun.display()
````

Create a Python class called `RectangularCoordinates` which represents an order pair that is on a Euclidean rectangular plane in form of $(x,y)$ point. Create the following methods:
- Create `__init__` method initialize with parameter $x$ and $y$
- A method to return the tuple of the position of this point.
- A method to check whether this point is on the $x$-axis or $y$-axis or not
- A method to find the quadrant (integer from 1 to 4) of this point by using the previous method for help
- Distance of this point to origin $(0,0)$
- A method `calculate_distance()` that accepts another `RectangularCoordinates` instance, and calculate the distance between those points
- A method to calculate the angle $\theta$ that this point does when draw a line to the origin $(0,0)$ with respect to the $x$-axis (The angle should be between $0^\circ\leq \theta \leq 90^\circ$)

````python
import math

class RectangularCoordinates:
    
    def __init__(self, x_plane, y_plane):
        self.x_plane = x_plane
        self.y_plane = y_plane

    def return_position(self):
        return (self.x_plane, self.y_plane)
    
    def check_axis(self) -> None:
        if self.x_plane == 0 and self.y_plane == 0 :
            return "in both x and y axis"
        elif self.y_plane == 0 :
            return "on X-axis"
        elif self.x_plane == 0 :
            return "on Y-axis"
        else:
            return "Not in axis"

    def find_quadrant(self):
        if self.x_plane > 0 and self.y_plane > 0 :
            return 1
        elif self.x_plane < 0 and self.y_plane > 0 :
            return 2
        elif self.x_plane < 0 and self.y_plane < 0 :
            return 3
        elif self.x_plane > 0 and self.y_plane < 0 :
            return 4
        else:
            return None
    
    def find_distance_to_origin(self):
        result = ((self.x_plane ** 2) + (self.y_plane ** 2)) ** 0.5
        return result
    
    def calculate_distance(self, another_point):
        different_x = another_point.x_plane - self.x_plane
        different_y = another_point.y_plane - self.y_plane
        distance = ((different_x ** 2) + (different_y ** 2)) ** 0.5
        return distance
    
    def calculate_angle(self):
        if self.x_plane == 0 :
            return 90 
        angle_radius = math.atan(abs(self.y_plane / self.x_plane))
        angle_degree = math.degrees(angle_radius)
        return angle_degree

wawa_in_point1 = RectangularCoordinates(1,1)
wawa_in_point2 = RectangularCoordinates(1,1)

print(wawa_in_point1.return_position())
print(wawa_in_point1.check_axis())
print(wawa_in_point1.find_quadrant())
print(wawa_in_point1.find_distance_to_origin())
print(wawa_in_point1.calculate_distance(wawa_in_point2))
print(wawa_in_point1.calculate_angle())
````
