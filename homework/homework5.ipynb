1. Write a decorator `n_times` which given a number `n` makes a function to execute `n` times.

```python
@n_times(3)
def print_my_name():
    print('Due')

print_my_name()
```

```terminal
Due
Due
Due
```

````python
# Your code here
import functools
def n_time(number):
    def decorator(func):
        @functools.wraps(func)
        def inner(*args, **kwargs):
            for _ in range(number):
                func(*args, **kwargs)
        return inner
    return decorator

@n_time(3)
def print_my_name() -> None:
    print("Due")

print_my_name()
````

2. Write a decorator `once` which makes a function to execute only once and other times it just returns the result calculated on the first call.

```python
@once
def random_number():
    return random.randint(1, 10)

print(random_number())
print(random_number())
print(random_number())
```

```terminal
4
4
4
```

````python
# Your code here
import random
import functools

def once(func):

    result = None

    @functools.wraps(func)
    def inner(*args, **kwargs):
        nonlocal result
        if not inner.called:
            result = func(*args, **kwargs)
            inner.called = True
        
        return result


    inner.called = False

    return inner

@once
def random_number() -> int:
    return random.randint(1, 10)

print(random_number())
print(random_number())
print(random_number())
````
