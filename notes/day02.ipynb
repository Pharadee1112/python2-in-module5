```python
address = input("Enter your address") # 123 Main Street Bangkok 10110
postal_code = address.split()[-1] 
# address.split() → ['123', 'Main', 'Street', 'Bangkok', '10110'] สมาชิกตัวสุดท้ายใน List จะเอาออกมาทำงาน

def print_postal_code(code):
    assert type(code) == str, f'Postal code need to be a string but get {type(code)}'
    assert len(code) == 5, f'Postal code need to be have 5 digit'
    print(code)

print_postal_code(postal_code)
```
