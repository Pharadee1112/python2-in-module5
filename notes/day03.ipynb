```python
class Cat:
    num_leg = 4 #มันคือ attribute ที่ติดอยู่กับ class 
    def __init__(self, name, sex, age, weight, texture, colur = "orange"):
        self.name = name
        self.sex = sex
        self.age = age
        self.weight = weight
        self.colur = colur
        self.texture = texture

    def greet(self): 
        print("meow meow")

    def print_info(self):
        print(f'Name : {self.name}\n Sex : {self.sex} \n Age : {self.age} \n Weight : {self.weight} \n Colur : {self.colur} \n Texture : {self.texture}\n Number of legs : {self.num_leg}')

    def get_bird_year(self):
        return 2025 - self.age
    
    def __str__(self):
        return f'The cat is named {self.name}. The cat aged {self.age}'

# oscar = Cat() #สร้างเเมว oscar ไม่ได้เเล้ว เพราะว่า เรากำหนดให้มันใช้ 6 argument นี้ แปลว่าต้องบังคัให้ใส่ข้อนี้ไปก่อน คือ
oscar = Cat(name = "Oscar", sex = "male", age = 3, weight = 7, colur = "brown", texture = "striped")
luna= Cat(name ='luna',sex = "male",age = 2,weight = 5,colur = "brown",texture = "striped")


oscar.print_info() 
# oscar.greet()
# oscar.get_bird_year()
```
