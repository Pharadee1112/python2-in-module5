```python
import time

def timethis(func):
    def wrapper(*args, **kwargs):
        start = time.time()                # เริ่มจับเวลา
        result = func(*args, **kwargs)    # เรียกฟังก์ชันจริง
        end = time.time()                  # จับเวลาหลังฟังก์ชันทำงานเสร็จ
        print(f"⏱ {func.__name__} ใช้เวลา {end - start:.5f} วินาที")
        return result
    return wrapper
```

```python
@timethis

def add(a, b):
    return a + b

add(2,5)
```

```python
import functools
from functools import wraps
```

```python
def timethis(func):
    @functools.wraps(func) # ฟังก์ชันที่เอาไว้เเทนฟังก์ชันเดิม ช่วยคัดลอกฟังก์ชันเดิมทั้งหมดเอาไว้ เอามาใช้กับ function inner

    def inner(*args, **kwargs):
        print(func.__name__, end="...")

        start_time = time.time()
        result = func(*args, **kwargs)
        elapse_time = time.time() - start_time

        print(elapse_time)
        return result
    return inner
```

## @once

เป็น decorator ที่ทำให้ ฟังก์ชันนั้นสามารถเรียกได้แค่ครั้งเดียว

ครั้งแรกที่เรียก → ฟังก์ชันทำงานตามปกติ

ครั้งต่อไป → ฟังก์ชัน ไม่ทำงาน

```python
def once(func):

    @functools.wraps(func)
    def inner(*args, **kwargs):

        if not inner.called:
            func(*args, **kwargs)
            inner.called = True

    inner.called = False 

    return inner
```

```python
@once
def f(x,y):
    return x+y

f(4,5)
```

```python
def once2(func):

    called = False

    @functools.wraps(func)
    def inner(*args, **kwargs):
        nonlocal called

        if not called:
            func(*args, **kwargs)
            inner.called = True

    # print(local())

    return inner
```
